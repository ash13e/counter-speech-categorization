```python
# all imported goods
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
hateReply = pd.read_csv("hate_reply.csv").set_index(["Reply_author"])
hateReply = hateReply.drop("Unnamed: 0", axis=1)
userInfo = pd.read_csv("user_info_curr.csv")
userInfo = userInfo.rename(columns={"author":"Reply_author"})
userInfo = userInfo.set_index("Reply_author")
userInfo = userInfo.drop("Unnamed: 0", axis=1)
total = pd.merge(hateReply, userInfo, on="Reply_author", how="left")
total.info()
total.columns
```

```python
ui2 = pd.read_csv("user_info_curr.csv").rename(columns={"author":"Hate_author"}).set_index("Hate_author").drop("Unnamed: 0", axis=1)
hr2 = pd.read_csv("hate_reply.csv").set_index("Hate_author").drop("Unnamed: 0", axis=1)
hr2 = hr2.drop(["Subreddit", "Submission", "Hate_body", "Hate_time", "Reply_author", "Reply_parentID", "Reply_body",
                     "Reply_output", "Reply_effectiveness"], axis=1)
total2 = pd.merge(hr2, ui2, on="Hate_author", how="left")
total2.info()
total2.columns
```

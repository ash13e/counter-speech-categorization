```python
# all imported goods
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
avi = pd.read_csv("data_reply_and_user_raw.csv")
```

```python
avi.info()
```

```python
avi = avi.drop(['Hate_author_y', 'Hate_ID_y'], axis=1)
avi = avi.rename(columns={'Hate_author_x':'Hate_author',
                          'Hate_ID_x': 'Hate_ID'})
```

```python
#old code to map year club
mapping = {'One-Year Club': 1,
           'Two-Year Club': 2,
           'Three-Year Club': 3,
           'Four-Year Club': 4,
           'Five-Year Club': 5,
           'Six-Year Club': 6,
           'Seven-Year Club': 7,
           'Eight-Year Club': 8,
           'Nine-Year Club': 9,
           'Ten-Year Club': 10,
           '11-Year Club': 11,
           '12-Year Club': 12,
           '13-Year Club': 13,
           '14-Year Club': 14,
           '15-Year Club': 15
          }
avi["author_Year_Club"] = avi["author_Year_Club"].replace(mapping)
#avi["author_Year_Club"] = pd.to_numeric(avi["author_Year_Club"])
avi["replier_Year_Club"] = avi["replier_Year_Club"].replace(mapping)
#avi["replier_Year_Club"] = pd.to_numeric(avi["replier_Year_Club"])
```

```python
#edited code to map gild
gMapping = {'Gilding I': 1,
            'Gilding II': 2,
            'Gilding III': 3,
            'Gilding IV': 4,
            'Gilding V': 5,
            'Gilding VI': 6,
            'Gilding VII': 7,
            'Gilding VIII': 8,
            'Gilding VIX': 9,
            'Gilding X': 10,
            'Gilding XI': 11,
            'Gilding XII': 12
           }
avi["author_Gilde"] = avi["author_Gilde"].replace(gMapping)
avi["replier_Gilde"] = avi["replier_Gilde"].replace(gMapping)
```

```python
avi.info()
avi.head(5)
```

```python
#turning account age into datetime
avi['author_create_date'] = pd.to_datetime(avi['author_create_utc'], unit ='s')
avi['replier_create_date'] = pd.to_datetime(avi['replier_create_utc'], unit ='s')
current_date = pd.to_datetime('now')
avi['author_age_of_account'] = (current_date - avi['author_create_date']).dt.days
avi['replier_age_of_account'] = (current_date - avi['replier_create_date']).dt.days
avi.info()
```

```python
avi.head(5)
```

```python
#using log distribution on Reply_effectiveness
epsilon = 1e-8
avi['log_Reply_effectiveness'] = np.log(avi['Reply_effectiveness'] + epsilon)
avi['log_author_comment_karma'] = np.log(avi['author_comment_karma'])
avi['log_replier_comment_karma'] = np.log(avi['replier_comment_karma'])
avi['log_author_link_karma'] = np.log(avi['author_link_karma'])
avi['log_replier_link_karma'] = np.log(avi['replier_link_karma'])
avi.info()
```
